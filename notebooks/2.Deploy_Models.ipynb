### 3. Inference with MLIS

#### Notes
- Using Kserve endpoint does not work with YOLO library, since it uses https and YOLO only accept http and grpc url for triton remote server. So we will use externalName of isvc's pod ( **e.g. http://[mlis deployment name]-predictor.[username namespace].svc.cluster.local** )
- if we deploy isvc in the namespaces which have requestauthentication and authorizationpolicy, we need to add token in the headers. ( *if we deploy isvc into the namespace which doens't have any requestauthentication and authorizationpolicy, we don't need to pass the AUTH_TOKEN in the headers.* )
- **Make sure below snippet is applied in the triton util file. if you patch followed [0.Introduction.ipynb](./0.Introduction.ipynb), then it's already patched.**

````python
vehicle_detector = YOLO("http://license-detectors-predictor.geuntak-6d6b6531.svc.cluster.local/vehicle_detector",task='detect')
license_detector = YOLO("http://license-detectors-predictor.geuntak-6d6b6531.svc.cluster.local/license_detector",task='detect')
````

````python
vehicle_results = vehicle_detector("./resources/Samples/car_image.jpg",save=True)
license_results = license_detector("./resources/Samples/car_image.jpg",save=True)
````

````python
import cv2 as cv
import matplotlib.pyplot as plt
````

````python
# cv.cvtColor(results[0].plot(), cv2.COLOR_BGR2RGB)
vehicle_rgb_img = cv.cvtColor(vehicle_results[0].plot(), cv.COLOR_BGR2RGB)
license_rgb_img = cv.cvtColor(license_results[0].plot(), cv.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(vehicle_rgb_img)
axes[1].imshow(license_rgb_img)
# plt.tight_layout()
plt.show()
````
